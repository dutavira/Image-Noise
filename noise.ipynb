## Import Library

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

## Import Image

```python
image = cv2.imread("D:/#KULIAH/SEMESTER 5/KECERDASAN BUATAN/Tugas/Image Noise/image/roger.jpg", 0)
```

```python
print(image.shape)
```

## Adding Noise

### Gaussian Noise

```python
gauss_noise = np.zeros((2600, 3400), dtype=np.uint8)
cv2.randn(gauss_noise, 100, 120)
gauss_noise = (gauss_noise*0.5).astype(np.uint8)
```

```python
gauss_img = cv2.add(image, gauss_noise)
```

```python
fig=plt.figure(dpi=300)

fig.add_subplot(1,3,1)
plt.imshow(image,cmap='gray')
plt.axis("off")
plt.title("Original")

fig.add_subplot(1,3,2)
plt.imshow(gauss_noise,cmap='gray')
plt.axis("off")
plt.title("Gaussian Noise")

fig.add_subplot(1,3,3)
plt.imshow(gauss_img,cmap='gray')
plt.axis("off")
plt.title("Combined")
```

## Removing Noise

### Non-local Means Denoising

```python
denoised1=cv2.fastNlMeansDenoising(gauss_img,None,50,50)

fig=plt.figure(dpi=300)
fig.add_subplot(1,3,1)
plt.imshow(image,cmap='gray')
plt.axis("off")
plt.title("Original")

fig.add_subplot(1,3,2)
plt.imshow(gauss_img,cmap='gray')
plt.axis("off")
plt.title("with Gaussian Noise")

fig.add_subplot(1,3,3)
plt.imshow(denoised1,cmap='gray')
plt.axis("off")
plt.title("After Denoising")
```
